## Gerar dados aleatórios e plotar um histograma (distribuição)
Gerar 30 valores (1 por dia) com minutos até 60.

```python
minutosPorDia = np.random.randint(0, 60, 30)
print(minutosPorDia)
```

Criação do histograma

```python
ax = sns.histplot(minutosPorDia)
ax.set(xlabel = 'Minutos Po Dia', ylabel = 'Qtd Vezes')
plt.show()
```

## Gerar uma matriz de correlação (de Pearson) de variáveis
Valor 1 é a correlação direta perfeita.  
Valor - 1 é a correlação negativa perfeita.  
Valores próximo do zero indicam a falta de existência ou dificuldade de estabelecer correlação.

```python
mask = np.triu(np.ones_like(dados.corr(numeric_only=True), dtype = np.bool_))
```

```python
plt.figure(figsize = (20, 20))
sns.heatmap(dados.corr(numeric_only=True), mask = mask, square = True, annot = True, vmin = -1, vmax = 1)
```
