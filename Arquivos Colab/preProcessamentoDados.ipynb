Verificar valores nulos

```python
dados.isnull()
```

Verificar valores nulos nas colunas

```python
dados.isnull().sum()
```

Coluna "winner_entry" tem muitos valores nulos e indica como o vencedor entrou no torneio.

```python
dados['winner_entry'].describe()
```

```python
dados['winner_entry'].unique()
```

Apagar todas as linhas que o atributo "loser_rank" é nulo, pois são poucas ocasiões e isso pode implicar erros nos resultados.

```python
dados.dropna(subset = ['loser_rank'], inplace = True)
print(dados)
```

Resultado: de 2733 linhas passou para 2707 linhas.

Apagar colunas que todos os valores são nulos.
obs: neste caso não há colunas com todos valores nulos.

```python
dados.dropna(axis = 1, how = 'all')
```

Alterar valor dos dados nulos em uma coluna.

```python
dados['loser_entry'].unique()
```

Substitui nulos por valor "x".
Utilizado especificamente para valores nulos.

```python
dados['loser_entry'].fillna(value = 'x', inplace = True)
```

```python
dados['loser_entry'].unique()
```

Alterar mais de um atributos, com condições diferentes.

```python
dados['w_ace'].unique()
```

```python
dados['loser_ht'].unique()
```

`'loser_ht': dados['loser_ht'].mean()` faz uma média dos valores que não são nulos e preenche com essa média.

```python
dados.fillna(value = {'w_ace': 0, 'loser_ht': dados['loser_ht'].mean()}, inplace = True)
```

```python
dados['w_ace'].unique()
```

```python
dados['loser_ht'].unique()
```

Usar o método `ffill` para projetar um dado para o registro seguinte.  
Se um dados é nulo, pega o dado anterior e coloca no dado nulo.

```python
dados['minutes'][1975:1990]
```

```python
dados['minutes'].fillna(method = 'ffill', inplace = True)
```

```python
dados['minutes'][1975:1990]
```

Identificar duplicados

```python
dados.duplicated()
```

```python
dados.duplicated().sum()
```

Eliminar duplicados de uma coluna (cuidado ao executar).

```python
dados.drop_duplicates(subset = ['tourney_name'], inplace = True)
```

```python
dados['tourney_name'].head(10)
```

```python
dados.describe()
```

# **Como fazer deslocamento**

Dados das ações da Apple
https://finance.yahoo.com/quote/AAPL/history/

```python
dados2 = pd.read_csv("AAPL.csv")
```

```python
dados2
```

```python
dados2['close_ontem'] = dados2['Close'].shift(1)
```

```python
dados2
```

Nova coluna com a porcentagem de alteração de um dia para o outro.

```python
dados2['alteracao %'] = ((dados2['Close']/dados2['close_ontem'])- 1)*100
```

```python
dados2.head(20)
```

```python
dados2.dropna(subset = ['alteracao %'], inplace = True)
```

```python
dados2
```

```python
dados2.describe(include = 'all')
```

# Tranformar os dados em categóricos(qualitativos) em dados numéricos.

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2021.csv')
```

País de origem do jogador.

```python
dados['winner_ioc']
```

Verificar valores únicos.

```python
dados['winner_ioc'].unique()
```

Ver documentação do scikit-learn sobre preprocessing  
https://scikit-learn.org/stable/auto_examples/index.html#preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
paises = dados['winner_ioc']
paises
```

Gerar países por números.

```python
lencoder = LabelEncoder()
paises_numeros = lencoder.fit_transform(paises)
paises_numeros
```

```python
paises.head(10)
```

Gerar uma nova coluna com o número dos países.

```python
dados['paises_numeros'] = paises_numeros
dados
```
